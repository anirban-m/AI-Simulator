```python
#Evolution of the atomic cloud in a multidimensional Mach Zender Interferometer
#over period of the experiment T_{exp} under a pulse sequence \pi/2-\pi-\pi/2 applied at times 0,T,2T
import numpy as np
import matplotlib.pyplot as plt
```

```python
Temp = 1.41*10**(-4)#Temperature in Kelvin
```

```python
m = 1.4*10**(-25)#atomic mass in kg
```

```python
kB = 1.38*10**(-23)#Boltzmann constant in m^{2} kg s^{-2} K^{-1} 
```

```python
hbar=6.626*10**(-34)
```

```python
sigmaVel = np.sqrt((2*kB*Temp)/m)#standard deviation for velocity
```

```python
x=np.linspace(-1.,1.,100)
y=np.linspace(-1.,1.,101)
z=np.linspace(-1.,1.,102)
X,Y,Z=np.meshgrid(x,y,z)
```

```python
gX = 9.9 #metres/sec^2 accerelation in x direction
```

```python
gY = 9.87 #metres/sec^2 accerelation in y direction
```

```python
gZ = 9.81 #metres/sec^2 accerelation in z direction
```

```python
Kx=(4*np.pi/780)*10**(9) #Momentum transfer due to Raman laser pulse along x direction
```

```python
Ky=(2*np.pi/780)*10**(9) #Momentum transfer due to Raman laser pulse along y direction
```

```python
Kz=(3*np.pi/780)*10**(9) #Momentum transfer due to Raman laser pulse along z direction
```

```python
T = 0.02 #Time between Raman laser pulses in seconds
```

```python
OmegaZ = 60/1e3
OmegaX = 50/1e3
OmegaY = 40/1e3 #rotational vector x,y,z components
```

```python
c=0.5 #contrast
Kz*gZ*T*T
```

```python
Texp=1 #experiment time in seconds 
DeltaPhiZX= Kz*gZ*T**2+Kx*(X/Texp)*(OmegaY-OmegaZ)*T**2-Kx*gX*T**2 +Kz*(Z/Texp)*(OmegaY-OmegaX)*T*T+(2*hbar/m)*Kx*Kz*OmegaY*T*T 
#Phase collected in quadrant-1 of the AI in the XZ plane
DeltaPhiXY= Kx*gX*T**2+Kx*(X/Texp)*(OmegaY-OmegaZ)*T**2-Ky*gY*T**2 +Ky*(Y/Texp)*(OmegaZ-OmegaX)*T*T+(2*hbar/m)*Kx*Ky*OmegaZ*T*T 
#Phase collected in quadrant-1 of the AI in the XY plane
DeltaPhiYZ= Ky*gY*T**2+Ky*(Y/Texp)*(OmegaZ-OmegaX)*T**2-Kz*gZ*T**2 +Kz*(Z/Texp)*(OmegaX-OmegaY)*T*T+(2*hbar/m)*Ky*Kz*OmegaZ*T*T 
#Phase collected in quadrant-1 of the AI in the XY plane
sigmaPos = sigmaVel*Texp #size of atomic cloud
fPSxz = (np.exp(-(X*X+Y*Y+Z*Z)/(2*sigmaPos**2))/((2*np.pi)**(3/2.)*sigmaPos**3))*(1+c*np.cos(DeltaPhiZX))*0.5 
#probabilistic distribution of the interferred atoms in the AI in the XZ plane
fPSxy = (np.exp(-(X*X+Y*Y+Z*Z)/(2*sigmaPos**2))/((2*np.pi)**(3/2.)*sigmaPos**3))*(1+c*np.cos(DeltaPhiXY))*0.5 
#probabilistic distribution of the interferred atoms in the AI in the XY plane
fPSyz = (np.exp(-(X*X+Y*Y+Z*Z)/(2*sigmaPos**2))/((2*np.pi)**(3/2.)*sigmaPos**3))*(1+c*np.cos(DeltaPhiYZ))*0.5 
#probabilistic distribution of the interferred atoms in the AI in the YZ plane
```

```python
fig,ax=plt.subplots(1,3,figsize=(14,4))
CS=ax[0].contourf(X[1,:],Z[1,:],fPSxz[1,:])
CS=ax[1].contourf(X[:,:,1],Y[:,:,1],fPSxy[:,:,1])
CS=ax[2].contourf(Y[:,1],Z[:,1],fPSyz[:,1])
ax[0].set_title(r"$\Phi_{x,z}$",fontsize=18)
ax[0].set_xlabel('x',fontsize=18)
ax[0].set_ylabel('z',fontsize=18)
ax[1].set_title(r"$\Phi_{x,y}$",fontsize=18)
ax[1].set_xlabel('x',fontsize=18)
ax[1].set_ylabel('y',fontsize=18)
ax[2].set_title(r"$\Phi_{y,z}$",fontsize=18)
ax[2].set_xlabel('y',fontsize=18)
ax[2].set_ylabel('z',fontsize=18)
cbar_ax = fig.add_axes([0.95, 0.15, 0.05, 0.7])
fig.colorbar(CS, cax=cbar_ax)
```
